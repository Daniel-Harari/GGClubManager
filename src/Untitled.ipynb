```python
import pandas as pd

def parse_excel_to_tables(file_path):
    # Load raw Excel data (no headers)
    df = pd.read_excel(file_path, header=None, sheet_name='Ring Game Detail')

    # Define a set of required table header keywords
    required_keywords = {'Player', 'Buy-in', 'Cashout', 'Hands'}

    # Create a mask of rows that contain all required header keywords
    mask = df.apply(lambda row: required_keywords.issubset(set(row.dropna().astype(str).str.strip())), axis=1)
    header_indices = df.index[mask].tolist()

    # Add the end of DataFrame as the final boundary
    header_indices.append(len(df))

    # Extract tables based on header locations
    tables = []
    for i in range(len(header_indices) - 1):
        start = header_indices[i]
        end = header_indices[i + 1]

        table = df.iloc[start:end].reset_index(drop=True)
        table.columns = table.iloc[0]  # First row is header
        table = table[1:].reset_index(drop=True)  # Drop header row

        tables.append(table)

    return tables
```
